1. **Rotação horária de 90°** e **shear (cisalhamento) em x**, diferentes dos exemplos apresentados em sala.


```
import numpy as np
import matplotlib.pyplot as plt
#matriz de transformação 90º e cisalhamento vertical
transformada = np.array([
    [0, 1],
    [-1, 1]
])

# 2. Defina o Vetor (v)
vetor = np.array([[2], [3]])

#Multiplicação de Matrizes
# O operador '@' executa a multiplicação matricial (T * v)
v_trans = transformada @ vetor




# --- GRÁFICO: Vetor Transformado ---
plt.figure(figsize=(6, 6))

#vetor original
plt.quiver(0, 0, vetor[0], vetor[1],
           angles='xy', scale_units='xy', scale=1,
           color='green', label='Vetor Original', width=0.015)

# Desenha o vetor transformado
# O vetor começa em (0, 0)
plt.quiver(0, 0, v_trans[0], v_trans[1],
           angles='xy', scale_units='xy', scale=1,
           color='purple', label='Vetor Transformado ', width=0.015)

# Configurações do gráfico
max_val = max(v_trans.flatten()) + 1 # Limite do eixo (neste caso, 3+1 = 4)
plt.title('Vetor Transformado ')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.xlim(0, max_val)
plt.ylim(0, max_val)
plt.grid()
plt.axhline(0, color='grey', lw=1)
plt.axvline(0, color='grey', lw=1)
plt.legend(loc='upper left')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

2. A obtenção das matrizes de transformação a partir dos vetores unitários (bases) — mostrando o processo numericamente e graficamente com o Matplotlib.

```
import numpy as np
import matplotlib.pyplot as plt

# ----------------------------------------------------------------------
# DEFINIÇÃO DOS VETORES BASE
# ----------------------------------------------------------------------
e1 = np.array([[1], [0]])
e2 = np.array([[0], [1]])
BASE_CANONICA = np.hstack([e1, e2])

# ----------------------------------------------------------------------
# GRÁFICO 1: BASE CANÔNICA ORIGINAL
# ----------------------------------------------------------------------
plt.figure(figsize=(6, 6))
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='blue', label='$e_1 = (1, 0)$', width=0.015)
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='cyan', label='$e_2 = (0, 1)$', width=0.015)

plt.title('Gráfico 1: Vetores Unitários da Base Canônica')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.grid()
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.legend(loc='upper right')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# ----------------------------------------------------------------------
# GRÁFICO 2: ROTAÇÃO HORÁRIA DE 90 GRAUS
# ----------------------------------------------------------------------
R = np.array([[0, 1], [-1, 0]])
e1_rot = R @ e1
e2_rot = R @ e2

print("\n--- ROTAÇÃO HORÁRIA DE 90° ---")
print(f"Matriz R:\n{R}")
print(f"T(e1) = R @ (1, 0) = {e1_rot.T[0]}")
print(f"T(e2) = R @ (0, 1) = {e2_rot.T[0]}")

plt.figure(figsize=(6, 6))
# Vetores Originais (para comparação)
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='lightgray', label='Original $e_1$', width=0.01)
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='lightgray', label='Original $e_2$', width=0.01)

# Vetores Transformados
plt.quiver(0, 0, e1_rot[0], e1_rot[1], angles='xy', scale_units='xy', scale=1, color='red', label='$e_1\'$ (0, -1)', width=0.015)
plt.quiver(0, 0, e2_rot[0], e2_rot[1], angles='xy', scale_units='xy', scale=1, color='orange', label='$e_2\'$ (1, 0)', width=0.015)

plt.title('Gráfico 2: Rotação Horária de 90 graus)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.grid()
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.legend(loc='upper right')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()

# ----------------------------------------------------------------------
# GRÁFICO 3: CISALHAMENTO VERTICAL
# ----------------------------------------------------------------------
C_vert = np.array([[1, 0], [1, 1]])
e1_shear = C_vert @ e1_rot
e2_shear = C_vert @ e2_rot

print("\n--- CISALHAMENTO VERTICAL ---")
print(f"Matriz C:\n{C_vert}")
print(f"T(e1) = C @ (1, 0) = {e1_shear.T[0]}")
print(f"T(e2) = C @ (0, 1) = {e2_shear.T[0]}")

plt.figure(figsize=(6, 6))
# Vetores Originais (para comparação)
plt.quiver(0, 0, e1[0], e1[1], angles='xy', scale_units='xy', scale=1, color='lightgray', label='Original $e_1$', width=0.01)
plt.quiver(0, 0, e2[0], e2[1], angles='xy', scale_units='xy', scale=1, color='lightgray', label='Original $e_2$', width=0.01)

# Vetores Transformados
plt.quiver(0, 0, e1_shear[0], e1_shear[1], angles='xy', scale_units='xy', scale=1, color='purple', label='$e_1\'\'$ (1, 1)', width=0.015)
plt.quiver(0, 0, e2_shear[0], e2_shear[1], angles='xy', scale_units='xy', scale=1, color='teal', label='$e_2\'\'$ (0, 1)', width=0.015)

plt.title('Gráfico 3: Cisalhamento Vertical ($C$ @ Base)')
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.grid()
plt.axhline(0, color='grey', lw=0.5)
plt.axvline(0, color='grey', lw=0.5)
plt.legend(loc='upper left')
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
